# Large matrices

```python
from time import time
import scipy.sparse
```

### Large NNZ: Timing a matrix with 20M nonzeros

```python
mat = gb.io.from_scipy_sparse(scipy.sparse.random(10_000, 10_000, density=0.2))
```

```python
# Measure spy plot creation time
start = time()

spy(mat, shading='absolute')

print(f"spy() runtime: {time()-start:.2}s.")
```

### Large dimensions: Timing a matrix with 10M rows and columns

```python
mat = gb.io.from_scipy_sparse(scipy.sparse.eye(10_000_000))
```

```python
# Measure spy plot creation time
start = time()

spy(mat)

print(f"spy() runtime: {time()-start:.2}s.")
```
