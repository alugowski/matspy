# Large matrices

Large matrices are no problem. A spy plot for a matrix with 20M elements takes less than half a second. Printing large matrices is as instantaneous as small ones.

```python
from time import time
```

### Timing a matrix with 20M nonzeros

```python
mat = scipy.sparse.random(10_000, 10_000, density=0.2).tocsr()
```

```python
# Measure spy plot creation time
start = time()

spy(mat, shading='absolute')

print(f"spy() runtime: {time()-start:.2}s.")

# Measure html pretty-print time
start = time()

to_html(mat)

print(f"to_html() runtime: {time()-start:.2}s.")
```
