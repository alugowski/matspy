# Precision

Sometimes we may wish to set near-zero values to zero. The `precision` argument does that. Any value `abs(value) < precision` is treated as zero.

This argument is compatible with [matplotlib.pyplot.spy()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.spy.html)'s `precision` parameter.

As a simple demonstration, use `precision` to filter random values:

```python
arr = np.random.random((100, 100))

from IPython.display import display, HTML
from matspy import to_sparkline

precisions = [0, 0.2, 0.8]
display(HTML(f'<table><tr>'
             f'<th style="text-align: center">precision = {precisions[0]}</th>'
             f'<th style="text-align: center">precision = {precisions[1]}</th>'
             f'<th style="text-align: center">precision = {precisions[2]}</th></tr><tr>'
             f"<td>{to_sparkline(arr, sparkline_size=1.5, precision=precisions[0])}</td>"
             f"<td>{to_sparkline(arr, sparkline_size=1.5, precision=precisions[1])}</td>"
             f"<td>{to_sparkline(arr, sparkline_size=1.5, precision=precisions[2])}</td>"
             f"</tr></table>"))
```
